##  <mark><span  style= "font-family: Times New Roman">Import necessary libraries</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##  <mark><span  style= "font-family: Times New Roman">View the data</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

```python


data = {
    "Batsman 1 score": [56, 19, 0, 68, 16, 67, 85, 101, 78],
    "Batsman 2 score": [60, 58, 38, 71, 69, 66, 68, 41, 67],
    "Temperature": [31, 33, 30, 35, 34, 15, 10, 7, 11]
}

df = pd.DataFrame(data)

print(df)
```


##  <mark><span  style= "font-family: Times New Roman">Data preprocessing</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

<div class="warning" style='padding:0.1em; background-color:#E9D8FD; color:#69337A'>
<span>
<p style='margin-top:1em; text-align:left'>
</p>
<p style='margin-left:1em;'>
</p><br>
   
  1. Third match score of Batsman-1 was mistakenly noted as 0. Kindly replace the value with the mean of earlier matches as it is not known.Round the calculated mean to the nearest integer. <br>    

2. Create a new feature in the dataframe naming 'weather' <br>
    
   a. Column should have 'Warm' if the temperature on the match day is greater than the mean of the temperatures.  <br>
   b. Column should have 'Cool' if the temperature on the match day is lesser than the mean of the temperatures.  <br>
</span>
</div>


```python
# Calculate the mean of earlier matches for Batsman-1
mean_batsman_1 = df.loc[df.index < 2, 'Batsman 1 score'].mean()

# Replace the mistaken value with the rounded mean
df.loc[df.index == 2, 'Batsman 1 score'] = round(mean_batsman_1)

print(df)
```

```python
mean_temperature = df['Temperature'].mean()

# Create 'weather' column based on temperature conditions
df['weather'] = df['Temperature'].apply(lambda x: 'Warm' if x > mean_temperature else 'Cool')

print(df)
```

##  <mark><span  style= "font-family: Times New Roman">Perform EDA</span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

### Visualize the evidences 

```python
print(df.describe())
```

```python
sns.pairplot(df)
plt.show()

```

```python
'''1. Positive correlations: None of the numerical columns seem to have a strong positive correlation.
   2. Distribution: The distribution of scores for both batsmen appears to vary.
```

```python
# Boxplot for Batsman 1 and Batsman 2 scores
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Batsman 1 score', 'Batsman 2 score']])
plt.title('Boxplot of Batsman 1 and Batsman 2 scores')
plt.tight_layout()  # Add this line to handle the warning
plt.show()
```

```python
'''1. Batsman 1: The scores vary, with a notable outlier in the third match (replaced with the mean).
   2. Batsman 2: Scores are relatively consistent without significant outliers.
```

```python
# Countplot for the 'weather' column
plt.figure(figsize=(6, 4))
sns.countplot(x='weather', data=df)
plt.title('Count of Matches by Weather Type')
plt.tight_layout()  # Add this line to handle the warning
plt.show()
```

```python
#There are more matches with 'Cool' weather than 'Warm' weathe
```
